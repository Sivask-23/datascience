```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
col={"red","cyan","pink","black","yellow"}
course=["java","c++","c","java","python"]
pop=[25,27,30,45,15]
plt.bar(course,pop,color=col,width=0.4)
plt.title("Popularity of Languages")
plt.xlabel("Languages")
plt.ylabel("Users")
plt.show()
```
