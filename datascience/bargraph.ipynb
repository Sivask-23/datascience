```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(4)
Y = np.array([5,1,3,10])
y1 = [3,4,2,8]
x_labels = ["G1","G2","G3","G4"]
```

```python
plt.bar(X-0.1,Y,color="hotpink",label="Men",width=0.2)
plt.bar(X+0.1,y1,color="red",label="Women",width=0.2)
plt.xticks(X,x_labels)
plt.xlabel('ALPHA')
plt.ylabel('RANGE')
plt.title('BAR GRAPH')
plt.legend()
plt.show()
```
