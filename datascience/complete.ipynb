```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acs,confusion_matrix,classification_report as cr
from sklearn.neighbors import KNeighborsClassifier
```

```python
# knn

iris=datasets.load_iris()
x=iris.data
y=iris.target
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
Y_pred=knn.predict(X_test)
print("Acccuracy: ",acs(y_test,Y_pred))
```

```python
# Naive Bayes


from sklearn.naive_bayes import GaussianNB,BernoulliNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
Y_pred=gnb.predict(X_test)
print("Acccuracy of Gaussian: ",acs(y_test,Y_pred))
bn=BernoulliNB()
bn.fit(X_train,y_train)
Y_pred=bn.predict(X_test)
print("Acccuracy of Bernoulli: ",acs(y_test,Y_pred))
```

```python
# Kmeans

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
x=iris.data[:,:2]
y=iris.target
km=KMeans(n_clusters=3,n_init=10,max_iter=500)
km.fit(x)
cent=km.cluster_centers_
newlab=km.labels_
fig,axes=plt.subplots(1,2,figsize=(16,8))
axes[0].scatter(x[:,0],x[:,1],c=y,cmap="winter",s=75)
axes[1].scatter(x[:,0],x[:,1],c=newlab,cmap="jet",s=75)
axes[1].scatter(cent[:,0],cent[:,1],marker="X",c="blue")
plt.show()
```
